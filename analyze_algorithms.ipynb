```python
import matplotlib.pyplot as plt
import numpy as np

# Data for visualization
algorithms = ['Logistic Regression', 'Random Forest', 'Multinomial Naive Bayes', 'Bernoulli Naive Bayes']
accuracy = [0.7783, 0.7178, 0.7773, 0.7640]
precision = [0.7788, 0.72, 0.775, 0.765]  # Average precision from classification reports
recall = [0.7783, 0.72, 0.775, 0.765]  # Average recall from classification reports
f1_score = [0.7782, 0.72, 0.775, 0.765]  # Average F1-score from classification reports
rmse = [0.4708, 0.5313, 0.4719, 0.4858]

# Plotting Accuracy, Precision, Recall, and F1-Score
plt.figure(figsize=(12, 6))
bar_width = 0.2
x = np.arange(len(algorithms))

# Create bars for each metric
plt.bar(x - bar_width * 1.5, accuracy, bar_width, label='Accuracy', color='blue')
plt.bar(x - bar_width * 0.5, precision, bar_width, label='Precision', color='green')
plt.bar(x + bar_width * 0.5, recall, bar_width, label='Recall', color='orange')
plt.bar(x + bar_width * 1.5, f1_score, bar_width, label='F1-Score', color='red')

# Adding labels and title
plt.title('Performance Metrics Comparison of Algorithms', fontsize=16)
plt.xlabel('Algorithms', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.xticks(x, algorithms, rotation=20, fontsize=10)
plt.ylim(0, 1)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save and Show the Plot
plt.tight_layout()
plt.savefig('performance_metrics_comparison.png')
plt.show()

# Plotting RMSE
plt.figure(figsize=(8, 5))
plt.bar(algorithms, rmse, color='purple', alpha=0.7)
plt.title('Root Mean Squared Error (RMSE) Comparison', fontsize=16)
plt.xlabel('Algorithms', fontsize=12)
plt.ylabel('RMSE', fontsize=12)
plt.xticks(fontsize=10)
plt.ylim(0, max(rmse) + 0.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save and Show the RMSE Plot
plt.tight_layout()
plt.savefig('rmse_comparison.png')
plt.show()
```
